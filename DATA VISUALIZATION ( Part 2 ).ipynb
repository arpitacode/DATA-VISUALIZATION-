# Draw histogram in python using matplotlib

1.	What is histogram:-

•	was first introduced by Karl Pearson.

•	It is graphical representation which organizes a group of data points into user-specific ranges.

•	A histogram shows the frequency on the vertical axis and bins in the horizontal axis.

•	The "bin" (or "bucket") is the range of values—that is, divide the entire range of values into a series of intervals—and then count how many values fall into each interval.

# ACTIVITY 1 -  Drawing simple histogram (Taking default bins=10)

```python
import matplotlib.pyplot as plt
marks=[40,60,55,20,35,70,60,89,20,33]
plt.hist(marks) 
plt.xlabel("Intervels(bins)")
plt.ylabel("Frequency of student") 
plt.title("Student Marks Distribution") 
plt.show()

```

# ACTIVITY 2 - Drawing simple histogram (Taking bins=5)

# ACTIVITY 4 Drawing simple histogram (Taking user defined 3 bins)


```python
import matplotlib.pyplot as plt
boy=[40,60,55,20,35,70,60,89,20,33]
plt.hist(boy,bins=[0,33,45,60,100])
plt.xlabel("Intervels(bins)") 
plt.ylabel("Frequency of student")
plt.title("Student Marks Distribution")

plt.show()
```

```python
import matplotlib.pyplot as plt
marks=[40,60,55,20,35,70,60,89,20,33]
plt.hist(marks,bins=[0,33,45,60,100],rwidth=0.9,color="green")
plt.xlabel("Intervels(bins)")
plt.ylabel("Frequency of student")
plt.title("Student Marks Distribution") 
plt.show()
```

# Activity 5 - How to change orientation 

```python
import matplotlib.pyplot as plt
marks=[40,60,55,20,35,70,60,89,20,33]
plt.hist(marks,bins=[0,33,45,60,100],rwidth=0.9,color="green",orientation ='horizontal')
plt.xlabel("Intervels(bins)")
plt.ylabel("Frequency of student")
plt.title("Student Marks Distribution") 
plt.show()
```

# ACTIVITY 6 - TWO data in Histogram 

```python
import matplotlib.pyplot as plt 
boys=[40,60,55,20,35,70,60,89,20,33]
girls=[50,68,44,90,32,15,86,74,54,68]
a=plt.hist([boys,girls],bins=[0,33,45,60,100],label=["Boys","Girls"],color=["red","green"],rwidth=0.9)
plt.xlabel("Intervels(bins)")
plt.ylabel("Frequency of student")
plt.title("Student Marks Distribution") 
plt.legend()
plt.show()
```

```python
ACTIVITY 7- HOW to convert graph into stacked 
```

```python
import matplotlib.pyplot as plt 
boys=[40,60,55,20,35,70,60,89,20,33]
girls=[50,68,44,90,32,15,86,74,54,68]
a=plt.hist([boys,girls],bins=[0,33,45,60,100],label=["Boys","Girls"],color=["red","green"],rwidth=0.9,stacked= True)
plt.xlabel("Intervels(bins)")
plt.ylabel("Frequency of student")
plt.title("Student Marks Distribution") 
plt.legend()
plt.show()
```
