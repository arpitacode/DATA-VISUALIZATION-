```python
plt.plot([1,2,3,4,5,6],[10,20,30,40,50,60])
plt.show()
```

# ACTIVITY - DIFFERENT FORMATS OF GRAPH

```python
import matplotlib.pyplot as plt
import numpy as np

a= np.arange(0,10,0.5)
a
```

```python
plt.plot(a,'go:')     # g for green, o for style is circle , : for dotted line  
```

```python
plt.plot(a,'go:',a*2,'rx:',a**2,'b+:')
plt.xlabel('X AXIS' )
plt.ylabel('y AXIS' )
plt.title('MY FIRST  COMBO GRAPH ')
plt.show()  
```

# GRID FUNCTION IN MATPLOT
Add Grid Lines to a Plot
With Pyplot, you can use the grid() function to add grid lines to the plot.# 

```python
# Display of grid lines 
plt.plot(a,'go:',a*2,'rx:',a**2,'b+:')
plt.xlabel('X AXIS' )
plt.ylabel('y AXIS' )
plt.title('MY FIRST  COMBO GRAPH ')
plt.grid()
plt.show()  
```

```python
# Change the colour of grid 
plt.plot(a,'go:',a*2,'rx:',a**2,'b+:')
plt.xlabel('X AXIS' )
plt.ylabel('y AXIS' )
plt.title('MY FIRST  COMBO GRAPH ')
plt.grid(color='red')
plt.show()  
```

# How To Create Bar chart 
The bar() function takes arguments that describes the layout of the bars.
The categories and their values represented by the first and second argument as arrays.

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["A", "B", "C", "D"]
y = [3, 8, 1, 10]

plt.bar(x,y)
plt.show()
```

```python
x = ["APPLES", "BANANAS"]
y = [400, 350]
plt.bar(x, y)
```

Horizontal Bars
If you want the bars to be displayed horizontally instead of vertically, use the barh() function:

```python
import matplotlib.pyplot as plt
import numpy as np

x = ["A", "B", "C", "D"]
y = [3, 8, 1, 10]

plt.barh(x, y)
plt.show()
```
